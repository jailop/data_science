Consulta a base de datos de publicaciones académicas.

```python
import json
import pandas as pd
from IPython.display import display, HTML
from jinja2 import Template
```

```python
with open('datasets/camjol/index.json', 'r') as fd:
    journals = json.load(fd)
```

```python
def find_in_title(word):
    arts = []
    for i, jrnl in enumerate(journals):
        for j, issue in enumerate(jrnl['issues']):
            for k, art in enumerate(issue['articles']):
                if art['title'].lower().find(word) >= 0:
                    arts.append(art)
    return arts
```

```python
def search(word):
    arts = find_in_title(word)
    with open('datasets/camjol/template.html', 'r') as fd:
        template = fd.read()
    t = Template(template)
    html = t.render(articles=arts)
    with open('datasets/camjol/output.html', 'w') as fd:
        fd.write(html)
    display(HTML(html))
```

```python
search('violencia')
```
